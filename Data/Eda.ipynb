```python
import pandas as pd

df = pd.read_csv("../Data/clean_patient_data.csv")
```

```python
print(df.head())
```

```python
# systolic clean
df["Systolic"] = df["Systolic"].str.replace("+","", regex=False)
df["Systolic"] = df["Systolic"].str.split("-").str[0].str.strip()
df["Systolic"] = df["Systolic"].astype(int)

# diastolic clean
df["Diastolic"] = df["Diastolic"].str.replace("+","", regex=False)
df["Diastolic"] = df["Diastolic"].str.split("-").str[0].str.strip()
df["Diastolic"] = df["Diastolic"].astype(int)
```

```python
df[["Systolic","Diastolic"]].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr = df[["Systolic","Diastolic"]].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("BP Correlation")
plt.show()
```

## Insights from Exploratory Data Analysis

- Most patients fall in the **age group 35-50**
- **Systolic and Diastolic blood pressure** show strong correlation
- Majority of patients are in **Stage-1 Hypertension**
